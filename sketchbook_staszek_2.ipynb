```python
class LinELM:
 
    def __init__(self, h=60):
        self.h = h
 
    def fit(self, X, y):
        self.b = LabelBinarizer()
        self.W = np.random.normal(size=X.shape[1] * self.h).reshape(X.shape[1], self.h)
        H = X.dot(self.W)
        self.b.fit(y)
        self.beta = la.pinv(H).dot(self.b.transform(y))
 
    def decision_function(self, X):
        return X.dot(self.W).dot(self.beta)
 
    def predict(self, X):
        return self.b.inverse_transform(self.decision_function(X))
 
class R2Lin:
 
    def __init__(self, k, estimator=partial(SVC, kernel='linear', C=1), recurrent=False, scale=False):
        self.k = k
        self.clf = estimator
        self.recurrent = recurrent
        self.activation = lambda x : 1.0 / (1.0 + np.exp(-x))
#        self.activation = lambda x : 1.0/np.sqrt(2*np.pi) * np.exp( - (x)**2  )
        self.alpha = 0.04
        self.scale = scale
 
    def fit(self, X, y):
        self.Wi = []
        self.clfs = []
        self.scalers = []
        self.C = len(set(y))
        Xi = X
        di = np.zeros(X.shape[0] * X.shape[1]).reshape(X.shape)
        for i in range(self.k):
 
            clf = self.clf()
            scaler = MinMaxScaler()
            if self.scale:
                Xi = scaler.fit_transform(Xi)
            self.scalers.append(scaler)
            clf.fit(Xi, y)
            self.clfs.append(clf)
            Oi = clf.decision_function(Xi)
            if self.C == 2:
                Oi = np.hstack([Oi,-Oi])
            Wi = np.random.normal(size=X.shape[1] * self.C).reshape(self.C, X.shape[1])
            self.Wi.append(Wi)
 
            if self.recurrent:
                di += Oi.dot(Wi)
            else:
                di = Oi.dot(Wi)
            Xi = self.activation(Xi + self.alpha * di)
        clf = self.clf()
        scaler = MinMaxScaler()
        if self.scale:
            Xi = scaler.fit_transform(Xi)
        self.scalers.append(scaler)
        clf.fit(Xi, y)
        self.clfs.append(clf)
        self.Wi.append(None)
        return self
 
    def predict(self, X):
        Xi = X
        di = np.zeros(X.shape[0] * X.shape[1]).reshape(X.shape)
        for clf, Wi, scaler in zip(self.clfs, self.Wi, self.scalers):
	 
            if self.scale:
                Xi = scaler.transform(Xi)
 
            if Wi is None:
                return clf.predict(Xi)
 
            Oi = clf.decision_function(Xi)
            if self.C == 2:
                Oi = np.hstack([Oi,-Oi])
 
            if self.recurrent:
                di += Oi.dot(Wi)
            else:
                di = Oi.dot(Wi)
            Xi = self.activation(Xi + self.alpha * di)
 
        
    
    def __str__(self):
        return 'R2Lin' + str(self.clf) + ' ' + str(self.k)
```

```python
class SVCWrapper(BaseEstimator):
    """ 
        This wrapper adapts sklearn model for R2SVMLearner 
        It returns always matrix of size n_samples x n_classes with real value predictions
    """
    
    def __init__(self, **params):
        self.model = sklearn.svm.SVC(**params)
    
    def fit(self, X, Y):
        self.model.fit(X, Y)
        
        if self.model.coef_.shape[0] == 1:
            self.coef_ = np.vstack([ self.model.coef_[0,:], -self.model.coef_[0,:]])
            self.intercept_ = np.array([self.model.intercept_[0], -self.model.intercept_[0]])
        else:
            self.coef_ = np.array(self.model.coef_)
            self.intercept_ = np.array(self.model.intercept_)
        
        return self
    
    def predict(self, X):
        return np.dot(X, self.coef_.T) + self.intercept_
    
    def set_params(self, **params):
        self.model.set_params(**params)
    
    def get_params(self, deep=True):
        return self.model.get_params(deep)
        #super(SVCWrapper, self).get_params(deep)
    
        
class R2SVMLearner(BaseEstimator):
    def __init__(self, beta=0.1, transf = 'tanh', seed=None, depth=5, recursive=True, base_cls=partial(SVCWrapper,
                                kernel='linear'), cls_params={}):
        self.seed = seed
        self.depth = depth
        self.recursive = recursive
        self.beta = beta
        self.models_ = []
        self.base_cls = base_cls
        self.scalers = []
        self.cls_params = cls_params
                
        if transf == 'tanh': 
            self.transf = lambda x: 2./(1.+np.exp(x)) - 1.
        elif transf == 'sigmoid':
            self.transf = lambda x: 1.0/(1.0 + np.exp(-x))
        else:
            self.transf = transf
        
        
    def fit(self, X, Y):
        assert(self.recursive is True)

        self.seed = self.seed if self.seed is not None else np.random.randint(0, np.iinfo(np.int32).max)
        self.random_state = np.random.RandomState(self.seed)

        self.models_ = []

        for i in xrange(self.depth): 
            self.models_.append(self.base_cls(**self.cls_params))  
            self.scalers.append(MinMaxScaler())
            self.models_[-1].set_params(random_state=self.random_state)
        
        self.models_[0].fit(X, Y)

        self.W, self.sigmas = [], None # self.depth - 1 sized array and concatenation
        self.X_tr = []
        
        X_mod = self.scalers[0].fit_transform(np.array(X))
        self.X_tr.append(X_mod)
        
        if not hasattr(self.models_[0], "coef_") or not hasattr(self.models_[0], "intercept_"):
            raise ValueError("base_cls should have coef_ and intercept_ fields") 
        
        # Main loop
        for i in range(self.depth - 1):            
            self.sigmas = np.hstack([self.sigmas, self.models_[i].predict(X_mod)]) \
                if i!=0 else self.models_[i].predict(X_mod)
            # Create random projection sized class number x number of features
            # Please note that we are doing transposition as opposed to paper
            
            self.W.append(self.random_state.normal(size=(self.sigmas.shape[1], X_mod.shape[1])))
            if i > 0:
                labels = self.sigmas.shape[1] / (i+1)
                self.W[-1][0:self.sigmas.shape[1]-labels,:] = self.W[-2]
                
            X_mod = self.scalers[i+1].fit_transform(self.transf(self.X_tr[-1] + self.beta*np.dot(self.sigmas, self.W[-1])))
            
            self.X_tr.append(X_mod)
            self.models_[i+1].fit(X_mod,Y)
            
 
        return self
    
    def score(self, X, Y):
        return sklearn.metrics.accuracy_score(Y, self.predict(X))
        
    def predict(self, X):
        return np.argmax(self._feed_forward(X), axis=1)
    
    def _feed_forward(self, X):
        ravel = False
    
        if len(X.shape) == 1:
            X = X.reshape(1, -1)
            ravel = True
            
        X_mod =self.scalers[0].transform(np.array(X))
        sigmas = None

        for i in range(self.depth - 1):
            sigmas = np.hstack([sigmas, self.models_[i].predict(X_mod)]) if i!=0 else self.models_[i].predict(X_mod)
            X_mod = self.scalers[i+1].transform(self.transf(X_mod + self.beta*np.dot(sigmas, self.W[i])))
            
        Y = self.models_[-1].predict(X_mod)
        
        return Y[0] if ravel else Y
    
    def get_params(self, deep=True):
        return {'beta': self.beta, 
                'transf': self.transf, 
                'depth': self.depth, 
                'recursive': self.recursive, 
                'base_cls': self.base_cls, 
                 'cls_params': self.cls_params}
        
```

```python
def grid_search(param_grid, X, Y, model=R2SVMLearner2, cv=3, max_iter=10, min_variance=0.15, verbose=True):
    
    grid_params = list(ParameterGrid(param_grid))
    scores = []
    params = []
    min_iter = max_iter/4
    
    def run(param) :
        score = 0.
        folds = KFold(len(X), cv, shuffle=True)
        for train_index, test_index in folds:
            temp_model = model(**param).fit(X[train_index], Y[train_index])
            score += sklearn.metrics.accuracy_score(Y[test_index], temp_model.predict(X[test_index]))
            
        return str(param), score/float(cv)
     
    for j, param in enumerate(grid_params) :
        p, s = run(param)
        params.append(p)
        scores.append(s)
        
    return params, scores
```
