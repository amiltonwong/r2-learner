```python
class SVCWrapper(BaseEstimator):
    """ 
        This wrapper adapts sklearn model for R2SVMLearner 
        It returns always matrix of size n_samples x n_classes with real value predictions
    """
    
    def __init__(self, **params):
        self.model = sklearn.svm.SVC(**params)
    
    def fit(self, X, Y):
        self.model.fit(X, Y)
        
        if self.model.coef_.shape[0] == 1:
            self.coef_ = np.vstack([ -self.model.coef_[0,:], self.model.coef_[0,:]])
            self.intercept_ = np.array([-self.model.intercept_[0], self.model.intercept_[0]])
        else:
            self.coef_ = np.array(self.model.coef_)
            self.intercept_ = np.array(self.model.intercept_)
        
        return self
    
    def predict(self, X):
        return np.dot(X, self.coef_.T) + self.intercept_
    
        
#TODO: Add special case for SVM and check for intercept / coef_ and use predict_proba in other cases
class R2SVMLearner(BaseEstimator):
    def __init__(self, beta=0.1, transf = 'sigmoid', depth=5, recursive=True, base_cls=SVCWrapper, 
                 cls_params={'kernel': 'linear'}):
        self.depth = depth
        self.recursive = recursive
        self.beta = beta
        self.models_ = []
        self.base_cls = base_cls
        self.cls_params = cls_params
        
        if transf == 'sigmoid': 
            self.transf = lambda x: 1.0/(1.0 + np.exp(-x))
        else:
            self.transf = transf
        
    def fit(self, X, Y):
        assert(self.recursive is True)
            
        #TODO: simplify code by np.empty
        
        self.W, sigmas = [], None # self.depth - 1 sized array and concatenation
        self.models_ = [self.base_cls(**self.cls_params).fit(X, Y)]
   
        self.X_tr = []
        
        X_mod = np.array(X)
        
        if not hasattr(self.models_[-1], "coef_") or not hasattr(self.models_[-1], "intercept_"):
            raise ValueError("base_cls should have coef_ and intercept_ fields") 
        
        # Main loop
        for i in range(self.depth - 1):            
            sigmas = np.hstack([sigmas, self.models_[-1].predict(X_mod)]) if i!=0 else self.models_[-1].predict(X_mod)
            # Create random projection sized class number x number of features
            # Please note that we are doing transposition as opposed to paper
            self.W.append(np.random.normal(0, 1, size=(sigmas.shape[1], X_mod.shape[1]))) 
            X_mod = self.transf(X + self.beta*np.dot(sigmas, self.W[-1]))
            
            self.X_tr.append(X_mod)
            
            self.models_.append(self.base_cls(**self.cls_params).fit(X_mod,Y))
  
        return self
    
    def score(self, X, Y):
        return sklearn.metrics.accuracy_score(Y, self.predict(X))
        
    def predict(self, X):
        return np.argmax(self._feed_forward(X), axis=1)
        #return (self._feed_forward(X) > 0.0).astype("int32")
    
    def _feed_forward(self, X):
        ravel = False
    
        if len(X.shape) == 1:
            X = X.reshape(1, -1)
            ravel = True
            
        X_mod = np.array(X)
        sigmas = None
        
        for i in range(self.depth - 1):
            sigmas = np.hstack([sigmas, self.models_[i].predict(X_mod)]) if i!=0 else self.models_[-1].predict(X_mod)
            X_mod = self.transf(X + self.beta*np.dot(sigmas, self.W[i]))
            
        Y = self.models_[-1].predict(X_mod)
        
        return Y[0] if ravel else Y
```

```python
X, Y = X_moon, Y_moon
model = R2SVMLearner(depth=50, beta=0.1, cls_params={'C': 10, 'kernel': 'linear'})
model.fit(X, Y)
print model.W[0]
%matplotlib inline
import matplotlib.pylab as plt
plt.scatter(model.X_tr[-1][:,0], model.X_tr[-1][:,1], c=Y)
plt.show()
print model.score(X,Y)
```

```python
def run_and_plot(X, Y, max_depth=4, beta=0.1, r2learn_params = {'C': 1, 'kernel': 'linear'}):
    plot_side = int(math.sqrt(max_depth))
    if plot_side**2 < max_depth: plot_side += 1

    for i in range(max_depth):
        model = R2SVMLearner(beta=beta, depth=i+1, cls_params=r2learn_params)
        model.fit(X, Y)
        
        plt.subplot(plot_side, plot_side, i + 1)
        
        N = 200

        x, y = np.linspace(X.min(), X.max(), N),  np.linspace(X.min(), X.max(), N)
        xx, yy = np.meshgrid(x, y)
        z = model.predict(np.c_[xx.ravel(), yy.ravel()])
        plt.contourf(xx, yy, z.reshape(N,N), cmap=plt.cm.Paired, alpha=0.8)
        plt.scatter(X[:,0], X[:,1], c=Y,  cmap=plt.cm.Paired, alpha=0.8)
        plt.show()
        
        
```

```python
run_and_plot(X_moon, Y_moon, beta=0.01, max_depth=2, r2learn_params = {'kernel': 'linear'})
```
