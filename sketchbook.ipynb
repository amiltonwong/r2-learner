```python
#TODO: use partial

class SVCWrapper(BaseEstimator):
    """ 
        This wrapper adapts sklearn model for R2SVMLearner 
        It returns always matrix of size n_samples x n_classes with real value predictions
    """
    
    def __init__(self, **params):
        self.model = sklearn.svm.SVC(**params)
    
    def fit(self, X, Y):
        self.model.fit(X, Y)
        
        if self.model.coef_.shape[0] == 1:
            self.coef_ = np.vstack([ -self.model.coef_[0,:], self.model.coef_[0,:]])
            self.intercept_ = np.array([-self.model.intercept_[0], self.model.intercept_[0]])
        else:
            self.coef_ = np.array(self.model.coef_)
            self.intercept_ = np.array(self.model.intercept_)
        
        return self
    
    def predict(self, X):
        return np.dot(X, self.coef_.T) + self.intercept_
    
    def set_params(self, **params):
        self.model.set_params(**params)
    
    def get_params(self, deep=True):
        return self.model.get_params(deep)
        #super(SVCWrapper, self).get_params(deep)
    
        
class R2SVMLearner(BaseEstimator):
    def __init__(self, beta=0.1, transf = 'tanh', depth=5, recursive=True, base_cls=SVCWrapper, 
                 seed = None,
                 cls_params={'kernel': 'linear', 'class_weight':'auto'}):
        self.seed = seed
        self.depth = depth
        self.recursive = recursive
        self.beta = beta
        self.models_ = []
        self.base_cls = base_cls
        self.cls_params = cls_params
                
        if transf == 'tanh': 
            #self.transf = lambda x: 1.0/(1.0 + np.exp(-x))
            self.transf = lambda x: 2./(1.+np.exp(x)) - 1.
        else:
            self.transf = transf
        
        
    def fit(self, X, Y):
        assert(self.recursive is True)

        self.seed = self.seed if self.seed is not None else np.random.randint(0, np.iinfo(np.int32).max)
        self.random_state = np.random.RandomState(self.seed)

        self.models_ = []

        for i in xrange(self.depth): 
            self.models_.append(self.base_cls(**self.cls_params))  
            self.models_[-1].set_params(random_state=self.random_state)
        
        self.models_[0].fit(X, Y)

        self.W, self.sigmas = [], None # self.depth - 1 sized array and concatenation
        self.X_tr = []
        
        X_mod = np.array(X)
        
        if not hasattr(self.models_[0], "coef_") or not hasattr(self.models_[0], "intercept_"):
            raise ValueError("base_cls should have coef_ and intercept_ fields") 
        
        # Main loop
        for i in range(self.depth - 1):            
            self.sigmas = np.hstack([self.sigmas, self.models_[i].predict(X_mod)]) \
                if i!=0 else self.models_[i].predict(X_mod)
            # Create random projection sized class number x number of features
            # Please note that we are doing transposition as opposed to paper
            
            self.W.append(self.random_state.normal(size=(self.sigmas.shape[1], X_mod.shape[1])))
            X_mod = self.transf(X + self.beta*np.dot(self.sigmas, self.W[-1]))
            
            self.X_tr.append(X_mod)
            
            #self.models_.append(self.base_cls(**self.cls_params).fit(X_mod,Y))
            self.models_[i+1].fit(X_mod,Y)
            
 
        return self
    
    def score(self, X, Y):
        return sklearn.metrics.accuracy_score(Y, self.predict(X))
        
    def predict(self, X):
        return np.argmax(self._feed_forward(X), axis=1)
        #return (self._feed_forward(X) > 0.0).astype("int32")
    
    def _feed_forward(self, X):
        ravel = False
    
        if len(X.shape) == 1:
            X = X.reshape(1, -1)
            ravel = True
            
        X_mod = np.array(X)
        sigmas = None
        
        for i in range(self.depth - 1):
            sigmas = np.hstack([sigmas, self.models_[i].predict(X_mod)]) if i!=0 else self.models_[i].predict(X_mod)
            X_mod = self.transf(X + self.beta*np.dot(sigmas, self.W[i]))
            
        Y = self.models_[-1].predict(X_mod)
        
        return Y[0] if ravel else Y
    
    def get_params(self, deep=True):
        return {'beta': self.beta, 
                'transf': self.transf, 
                'depth': self.depth, 
                'recursive': self.recursive, 
                'base_cls': self.base_cls, 
                 'cls_params': self.cls_params}
        
```

```python
class R2SVM():
    def __init__(self,l=2,C=1,beta=0.1, seed=None):
        self.seed = seed
        self.beta = beta
        self.layers = []
        self.outputs = []
        self.depth = l
        self.C = C
        self.X_tr = []

    def sigmoid(self,x):
        return 2/(1+np.exp(x)) - 1
        
    def get_input(self, x, layer, o ):
 
        if layer == 1: return x
        else:
            if o == None:
                o = self.get_output(x, 1, None)
                for previous in xrange(layer-2):
                    o = np.concatenate((o,self.get_output(x, previous+2,o)),1)
            move = ( np.array(o).dot( np.array(self.projections[layer-2])))
            x_propagated = self.sigmoid(x+self.beta*move)
 
        return x_propagated 
 
    def get_output(self, x, layer, o = None):
        if layer != 1 and o == None: 
            o = self.get_output(x, 1, None)
            for previous in xrange(layer-2):
                o = np.concatenate((o,self.get_output(x, previous+2,o)),1)
        inp = self.get_input(x, layer, o)
        out = np.round( self.layers[layer-1].decision_function(inp),2)
        if (1==out.shape[1]): 
            tmp_out = []
            for x in out:
                tmp_out.append([-x[0],x[0]])
            out = tmp_out
        return out
 
    def fit(self,x,y, projections=[]):
        self.seed = self.seed if self.seed is not None else np.random.randint(0, np.iinfo(np.int32).max)
        self.random_state = np.random.RandomState(self.seed)
        self.projections = projections
        self.X_tr = []
        self.layers = []
        
        for i in xrange(self.depth): 
            self.layers.append(svm.SVC(C=self.C,kernel='linear',class_weight='auto'))  
        
        for m in self.layers:
            m.set_params(random_state=self.random_state)

        classes = len(np.unique(y))
        features = len(x[0])
   

        for i in xrange(len(self.layers)):
            
            if i==0:
                self.layers[0].fit(x,y)
                o = self.get_output(x, 1, None)      
                if len(self.projections) != len(self.layers)-1:
                    self.projections.append(self.random_state.normal(size=(classes*(i+1),features)))
                continue            
            
            
            if len(self.projections) != len(self.layers)-1:
                self.projections.append(self.random_state.normal(size=(classes*(i+1),features)))
            self.layers[i].fit( self.get_input(x, i+1, o), y)
            
            self.X_tr.append(self.get_input(x, i+1, o))
            o = np.concatenate((o,self.get_output(x, i+1, o)),1)
 
        self.outputs = o
            
        return self
 
    def classify_internal(self,x,layer):
        return [np.argmax(p) for p in self.get_output( x, layer )]
 
    def score(self, X, Y):
        return sklearn.metrics.accuracy_score(Y, self.predict(X))
    
    def predict(self,x): 
        return self.classify_internal( x, len(self.layers) )
 
    def drawdata(self, x, y):
        _x = np.array(self.get_input(x,len(self.layers),None))
        plt.scatter(_x[:,0].tolist(),_x[:,1].tolist(),c=y)
        plt.show()
        
    def score(self, X, Y):
        return sklearn.metrics.accuracy_score(Y, self.predict(X))
```

```python
def run_and_plot(X, Y, max_depth=4, beta=0.1, r2learn_params = {'C': 1, 'kernel': 'linear'}):
    plot_side = int(math.sqrt(max_depth))
    if plot_side**2 < max_depth: plot_side += 1

    for i in range(max_depth):
        model = R2SVMLearner(beta=beta, depth=i+1, cls_params=r2learn_params)
        model.fit(X, Y)
        
        plt.subplot(plot_side, plot_side, i + 1)
        
        N = 200

        x, y = np.linspace(X.min(), X.max(), N),  np.linspace(X.min(), X.max(), N)
        xx, yy = np.meshgrid(x, y)
        z = model.predict(np.c_[xx.ravel(), yy.ravel()])
        plt.contourf(xx, yy, z.reshape(N,N), cmap=plt.cm.Paired, alpha=0.8)
        plt.scatter(X[:,0], X[:,1], c=Y,  cmap=plt.cm.Paired, alpha=0.8)
        plt.show()
        
        
```

```python
def grid_search(param_grid, X, Y, model=R2SVMLearner, cv=3, max_iter=10, min_variance=0.15, verbose=True):
    
    grid_params = list(ParameterGrid(param_grid))
    scores = []
    params = []
    min_iter = max_iter/4
    
    def run(param) :
        score = 0.
        folds = KFold(len(X), cv, shuffle=True)
        for train_index, test_index in folds:
            temp_model = model(**param).fit(X[train_index], Y[train_index])
            score += sklearn.metrics.accuracy_score(Y[test_index], temp_model.predict(X[test_index]))
            
        return str(param), score/float(cv)
     
    for j, param in enumerate(grid_params) :
        p, s = run(param)
        params.append(p)
        scores.append(s)
        
#         t_scores = []
#         for i in xrange(max_iter):
#             # print "%i/%i - %i/%i" % (j, len(grid_params)-1, i, max_iter-1)
#             p, s = run(param)
#             t_scores.append(s)
#             if i > min_iter and np.std(t_scores) <= min_variance :
#                 params.append(p)
#                 scores.append(s)
#                 break
#             if i == max_iter - 1 and verbose:
#                 print "No convergence for %s %i/%i" % (str(param), j, len(grid_params))
#                 print "Score variance:", np.std(t_scores)

    return params, scores

X, Y = X_moon, Y_moon
param_grid = {'depth': [5], 'beta': [0.1, 0.2], 'cls_params': [{'C': c, 'kernel': 'linear'} for c in [1,10]]}

params, scores = grid_search(param_grid, X, Y)
print params[np.argmax(scores)], np.max(scores)
```
