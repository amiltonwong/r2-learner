```python
def plot(model, X, Y, layers=2):
    X_tr = model._X_tr
    
    N = 200
    plot_size=(10, 40)
    x, y = np.linspace(X[:,0].min(), X[:,0].max(), N),  np.linspace(X[:,1].min(), X[:,1].max(), N)
    xx, yy = np.meshgrid(x, y)
    z = model.predict(np.hstack([np.c_[xx.ravel(), yy.ravel(), np.zeros(shape=(N*N,zeros_size))] for _ in range(repeat)]))

    
    fig = plt.figure(figsize=plot_size)
    ax1 = fig.add_subplot(layers+1,1,1)
    ax1.contourf(xx, yy, z.reshape(N,N), cmap=plt.cm.Paired, alpha=0.8) #levels=np.arange(-2,2,0.2),
    ax1.scatter(X[:,0], X[:,1], c=Y,  cmap=plt.cm.Paired, alpha=0.8)

    for i in reversed(range(layers)):
        ax2 = fig.add_subplot(layers+1,1,2+(i))

        if len(model.W) > i:
            ax2.plot([0, model.W[i][1,0] - model.W[i][0,0]], [0, model.W[i][1,1] -  model.W[i][0,1]], 'k-', lw=4, c='r')

        ax2.contourf(model._X_tr[-i-1][:,0].reshape(N,N), model._X_tr[-i-1][:,1].reshape(N,N), z.reshape(N,N), 
                     cmap=plt.cm.Paired, alpha=0.8) #levels=np.arange(-2,2,0.2), 
        ax2.scatter(X_tr[-i-1][:,0], X_tr[-i-1][:,1], c=Y)
    plt.show()
plot(model, X, Y, layers=model.depth)
```

```python
def multiply_dim(data, factor=10):
    return np.hstack([data for _ in xrange(factor)])
```

```python
spirals = fetch_synthetic_datasets().spiral
factor = 200
depth = 3
```

```python
normal_model = R2SVMLearner(seed=674, use_linear_svc=True, depth=3, beta=0.9, fit_c=None, use_prev=False, scale=True, 
                 activation='sigmoid')
multi_model = R2SVMLearner(seed=674, use_linear_svc=True, depth=3, beta=5., fit_c=None, use_prev=False, scale=True, 
                 activation='sigmoid')

cm_data = ListedColormap(['#FF0000', '#0000FF'])
cm_cont = plt.cm.RdBu

X = MinMaxScaler((-1,1)).fit_transform(spirals.data)
multi_X = multiply_dim(X, factor)
Y = spirals.target

normal_model.fit(X, Y)
multi_model.fit(multi_X, Y)

N = 200
x, y = np.linspace(X[:,0].min() - 0.1, X[:,0].max() + 0.1, N),  np.linspace(X[:,1].min() - 0.1, X[:,1].max() + 0.1, N)
xx, yy = np.meshgrid(x, y)

grid = np.c_[xx.ravel(), yy.ravel()]
multi_grid = np.hstack([np.c_[xx.ravel(), yy.ravel()] for _ in xrange(factor)])

predictions = normal_model.predict(grid, all_layers=True)
multi_predictions = multi_model.predict(multi_grid, all_layers=True)

for i, char in enumerate(['a','b','c']):
    fig = plt.figure(figsize=(10,10))
    
    ax = fig.add_subplot(111)
    ax.set_aspect(1)
    
    ax.contourf(grid[:,0].reshape(N,N), grid[:,1].reshape(N,N), 
               predictions[i].reshape(N,N), cmap=cm_cont, alpha=0.4,  antialiased=True)
    ax.scatter(X[:,0], X[:,1], c=Y, cmap=cm_data, alpha=0.8)
    ax.axis('off')
    
    plt.savefig('spiral_' + char, bbox_inches='tight')
    
for i, char in enumerate(['d','e','f']):
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111)
    ax.set_aspect(1)
    
    ax.contourf(grid[:,0].reshape(N,N), grid[:,1].reshape(N,N),
                multi_predictions[i].reshape(N,N), cmap=cm_cont, alpha=0.4,  antialiased=True)
    ax.scatter(X[:,0], X[:,1], c=Y, cmap=cm_data, alpha=0.8)
    ax.axis('off')
    
    plt.savefig('spiral_' + char + '.png', bbox_inches='tight')
```

```python
X = MinMaxScaler((-1,1)).fit_transform(moon.data)
Y = moon.target

i = 0;
m = 0
best_seed = 0
while True:
    mo = R2SVMLearner(seed=i, use_linear_svc=True, depth=3, beta=0.9, fit_c=None, use_prev=False, scale=True, 
                 activation='sigmoid')
    
    Y_pred = mo.fit(X,Y).predict(X)
    score = accuracy_score(Y, Y_pred)
    
    if score > m:
        m = score
        best_seed = i
    if abs(score - 1) < 0.001:
        m = score
        best_seed = i
        print "!!!!!!! found best score:", m, "with seed:", best_seed, i
        break
    
    if i % 1000 == 0:
        print "best score so far:", m, "with seed:", best_seed
    
    sys.stdout.flush()
    
    i += 1
```
