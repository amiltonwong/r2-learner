```python
def plot(model, X, Y, layers=2):
    X_tr = model._X_tr
    
    N = 200
    plot_size=(10, 40)
    x, y = np.linspace(X[:,0].min(), X[:,0].max(), N),  np.linspace(X[:,1].min(), X[:,1].max(), N)
    xx, yy = np.meshgrid(x, y)
    z = model.predict(np.hstack([np.c_[xx.ravel(), yy.ravel(), np.zeros(shape=(N*N,zeros_size))] for _ in range(repeat)]))

    
    fig = plt.figure(figsize=plot_size)
    ax1 = fig.add_subplot(layers+1,1,1)
    ax1.contourf(xx, yy, z.reshape(N,N), cmap=plt.cm.Paired, alpha=0.8) #levels=np.arange(-2,2,0.2),
    ax1.scatter(X[:,0], X[:,1], c=Y,  cmap=plt.cm.Paired, alpha=0.8)

    for i in reversed(range(layers)):
        ax2 = fig.add_subplot(layers+1,1,2+(i))

        if len(model.W) > i:
            ax2.plot([0, model.W[i][1,0] - model.W[i][0,0]], [0, model.W[i][1,1] -  model.W[i][0,1]], 'k-', lw=4, c='r')

        ax2.contourf(model._X_tr[-i-1][:,0].reshape(N,N), model._X_tr[-i-1][:,1].reshape(N,N), z.reshape(N,N), 
                     cmap=plt.cm.Paired, alpha=0.8) #levels=np.arange(-2,2,0.2), 
        ax2.scatter(X_tr[-i-1][:,0], X_tr[-i-1][:,1], c=Y)
    plt.show()
plot(model, X, Y, layers=model.depth)
```

```python
fig = plt.figure(figsize=(10,10))
X_tr = model._X_tr
for i in range(model.depth-1):
    ax2 = fig.add_subplot(1,model.depth,i+1)
    ax2.set_aspect(1)
    
    N = 200
    plot_size=(10, 10)
    x, y = np.linspace(X[:,0].min(), X[:,0].max(), N),  np.linspace(X[:,1].min(), X[:,1].max(), N)
    xx, yy = np.meshgrid(x, y)
    z = model.predict(np.hstack([np.c_[xx.ravel(), yy.ravel(), np.zeros(shape=(N*N,zeros_size))] for _ in range(repeat)]))

    
    ax2.plot([0, model.W[i][1,0] - model.W[i][0,0]], [0, model.W[i][1,1] -  model.W[i][0,1]], 'k-', lw=4, c='r')
    
    ax2.contourf(model._X_tr[i][:,0].reshape(N,N), model._X_tr[i][:,1].reshape(N,N), z.reshape(N,N), 
                 cmap=plt.cm.Paired, alpha=0.8) #levels=np.arange(-2,2,0.2), 
    ax2.scatter(X_tr[i][:,0], X_tr[i][:,1], c=Y)
plt.show()
```
